```python
# 부등호 사용하기 
print(10 > 20)
print(10 < 20)
print(10 <= 20)
print(10 <= 10)
print(10 >= 20)
```

```python
# 객체가 같은지 다른지 비교하기

print(1 == 1.0) # ==는 값이 같은지 다른지 비교하는 연산자이다. 1과 1.0은 값이 같기 때문에 True가 나온다.
print(1 is 1.0) #  is는 객체가 같은지 다른지 비교하는 연산자이다. 1과 1.0은 값은 같지만 객체는 다르다. 그래서 False가 나온다.
```

```python
# 논리 연산자 사용하기 
## AND
print(True and True)
print(True and False)
print(False and True)
print(False and False)

# and는 모든 경우가 True일 때만 True가 나옴.
```

```python
# OR
print(True or True)
print(True or False)
print(False or True)
print(False or False)
```
