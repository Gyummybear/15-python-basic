```python
# 변수 여러 개를 한 번에 만들기

x , y, z = 10, 20, 30
print(x)
print(y)
print(z)

```

```python
x, y = 10, 20
x, y = y, x # x 값과 y값이 서로 교차
print(x)
print(y)
```

```python
# 변수로 계산하기
a = 10
b = 20
# a = "10"
# b = "20"
## c = a + b 이렇게 하면 문자열 그대로 '1020'이 나옴

# a = "10"
# b = 20
## c = a + b 이렇게 하면 문자열과 숫자는 더할 수 없어서 오류가 난다


c = a + b
print(c)
```

```python
# 산술 연산 후 할당 연산자 사용하기
a = 10
a = a + 20
print(a) # 위의 코드로 a는 30으로 재할당 됨
```

```python
a = input("첫 번째 숫자 :")
b = input("두 번째 숫자 :")
# 이것도 수학적인 덧셈 형식이 아닌 문자열 그대로 출력 됨
print(a + b)
```

```python
# 입력 값을 정수로 변환하기

a = int(input("첫 번째 숫자 :"))
b = int(input("두 번째 숫자 :"))

print(a + b)
```

```python
# 압력 값을 변수 두 개에 저장하기

a ,b = input("문자열 두 개를 입력하세요").split() # split() 함수는 문자열을 공백 기준으로 나눠서 리스트 만들어줌

print(a)
print(b)
```

```python
# "hello python".split()

a, b = ['hello', 'python'] # 3개가 아니라 2개이므로 오류가 난다
print(a)
print(b)
```

```python
# 두 숫자의 합 구하기
a , b = input("숫자 2개를 입력하세요 : ").split()
print(a)
print(b)
print(a+b)
```

```python
# 두 숫자의 합 구하기
a, b = input("숫자 2개를 입력하세요 : ").split()
a = int(a)
b = int(b)
print(a+b)
```

```python
# 입력 받은 값을 콤마를 기준으로 분리하기

a,b = map(int, input("숫자 두 개를 입력하세요 : ").split(","))
# a,b = map(int, ["10,20"] 이렇게 하면 리스트 안에 문자열이 하나 있기 때문에 오류가 난다

# a,b = 10 ,20 

print(a + b)

```
