```python
# 값을 여러개 출력하기

print(1, 2, 3)
```

```python
# sep로 값 사이에 문자 넣기
print(1,2,3, sep = ", ") # 'Separator'의 약자입니다. 우리말로 하면 '구분자
```

```python
print(1920, 1000, sep = "x")
```

```python
# 줄 바꿈 활용하기 
print(1, 2, 3, sep = "\n") # \n은 줄 바꿈을 의미하는 특수 문자입니다
```

```python
print("1 \n2 \n3")  # sep를 안하더라도 \n을 이용해서 줄 바꿈을 할 수 있습니다.
```

```python
# end 사용하기
print(1)
print(2)
print(3)
```

```python
print(1, end = " ")  # end는 print() 함수는 기본적으로 출력이 끝난 뒤에 줄 바꿈을 합니다. 
print(2, end = " ") 
print(3) 
## 하지만 end 매개변수에 다른 문자를 넣으면 줄 바꿈 대신 그 문자가 출력됩니다. 위의 코드는 1과 2 사이에 공백이 들어가도록 하는 코드입니다.
```

```python
print(1, end = " ")
print(2, end = " ")
print(3)

print(1,2,3)
```

```python
year = 2000
month = 10 
day = 27
hour = 11
minute = 43
second = 59
print(year, month, day, sep = "/" , end = " ")
print(hour, minute, second, sep = ":")
```
