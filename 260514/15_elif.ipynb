```python
# 음료수 자판기

button = int(input())

if button == 1 :
    print("콜라")
elif button == 2:
    print("사이다")
elif button == 3:
    print("환타")
else:
    print("제공하지 않는 메뉴")
```

```python
x = int(input())

if 11 <= x <= 20 :
    print("11~20")
elif 21 <= x <= 30 :
    print("21~30")
else :
    print("아무것도 해당하지 않음")
```

```python
age = int(input())
balance = 9000

if 7 <= age <= 12 :
    balance -= 650 
    print(balance)

elif 13<= age <= 18 :
    balance -= 1050 
    print(balance)

elif 19 <= age :
    balance -= 1250 
    print(balance)

else :
    print("7세 이하는 무료입니다")
```
