```python
# 키 이름이 중복이 되면?
lux = {
    "health" : 490, 
    "mana" : 800,
    "melee" :550,
    "armor" :18.72
} 
print(lux) # lux값 다 가져옴
print(lux["health"]) # lux에 있는 "health" key의 value 값을 가져옴 
```

```python
x = {
    100 : "Hundred",
    False : 0,
    3.5 : [3.5, 3.5] # value 값은 list 가능
}

print(type(x))
print(x)
```

```python
y = {
    100 : "Hundred",
    False : 0,
    [3.5, 3.5] : 3.5 # key 값으로는 list 불가능
}

print(type(y))
print(y)
```

```python
# 빈 딕셔너리 만들기

x = {}
y = dict() # () 값을 딕셔너리로 바꾼다?
print(x)
print(bool(x)) # 비어 있으면 False
print(type(x))
print(y)
print(bool(y))
print(type(y))
```

```python
# dict 함수로 딕셔너리 만들기

lux1 = dict(health = 400, mane=334, melee = 50, armor = 10.72)
# dict(key = value,key = value,key = value, ....) -> key 값과 value 값 지정할 수 있음

print(type(lux1))
print(lux1)
```

```python
lux2 = dict(zip(["health","mane", "melee","armor"],[490,334,550,10.72]))
#zip -> key 값과 , value 값을 결합해주는 함수  # map 함수랑 비슷한가..?
# dict 형식으로 만드려면 zip 앞에다가 dict을 작성 해줘야함  
print(lux2)
```

```python
lux3 = dict([("health",490),("mane",334),("melee",550),("armor",10.72)])

print(lux3)
print(type(lux3))
```

```python
# 딕셔너리의 키에 값 할당하기 
lux = {
    "health" : 490, 
    "mana" : 800,
    "melee" :550,
    "armor" :18.72
} 

print(lux["health"]) # lux에서 "health" 값 출력
print(lux["armor"])

lux["health"] = 2037 # health값에 2037을 업데이트
print(lux)
```

```python
lux["mana_regen"] = 3.28 # 딕셔너리에서는 없는 key,value 추가 가능 , 리스트는 추가 불가능
print(lux) 
```

```python
lux["attack_speed"] # 없는 것은 찾을 수 없음
```

```python
# 딕셔너리에 키가 있는지 확인하기 
print(lux)
print("health" in lux) # lux에 health가 있는지 확인
print("attack_speed" in lux)
```

```python
# 딕셔너리의 키의 갯수 구하기

print(lux)
len(lux) # key , value 값 갯수확인  
```
