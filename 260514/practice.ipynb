```python
students = {
    "김철수": [90, 80, 70],
    "이영희": [100, 95, 90],
    "박민수": [60, 70, 65]
}

# 1. 이름들만 리스트로 쏙 뽑아두기 (가장 핵심!)
이름목록 = list(students.keys())

print("===== 학생 성적 결과 =====")

# --- 첫 번째 학생 (김철수) ---
이름1 = 이름목록[0]
점수1 = students[이름1]
평균1 = sum(점수1) / len(점수1)
print(f"이름: {이름1}")
print(f"점수: {점수1}")
print(f"평균: {평균1}")
print() # 한 줄 띄우기

# --- 두 번째 학생 (이영희) ---
이름2 = 이름목록[1]
점수2 = students[이름2]
평균2 = sum(점수2) / len(점수2)
print(f"이름: {이름2}")
print(f"점수: {점수2}")
print(f"평균: {평균2}")
print()

# --- 세 번째 학생 (박민수) ---
이름3 = 이름목록[2]
점수3 = students[이름3]
평균3 = sum(점수3) / len(점수3)
print(f"이름: {이름3}")
print(f"점수: {점수3}")
print(f"평균: {평균3}")

# --- 전체 결과 계산 ---
전체학생수 = len(students)
전체평균 = (평균1 + 평균2 + 평균3) / 전체학생수

print("===== 전체 결과 =====")
print(f"전체 학생 수: {전체학생수}명")
print(f"전체 평균: {전체평균}")
```
