```python
# while 조건문 :  # True False
#     반복할 코드

i = 0

while i < 100 :
    print("hello world")
    i += 1 # i에 1을 대입
```

```python
# 초기값을 1부터 시작하기

i = 1 
while i <= 10 :
    print("hello world")
    i += 1
```

```python
# 초기값을 감소 시키기

i = 10

while i > 0:
    print("hello world", i)
    i -=1
```

```python
# 입력한 횟수대로 반복하기

count = int(input("반복할 횟수를 입력하세요"))

i = 0 

while i < count :
    print("hello world", i)
    i += 1
```
