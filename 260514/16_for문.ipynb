```python
# for와 range 사용하기 
# for 변수 in range(횟수): # 변수에 있는지 없는지 확인하고 횟수 만큼 돌리기
#     반복할 코드  

for i in range(10) : # i는 왜 숫자가 들어가지..
    print(i)

list(range(10))
```

```python
# 반복문에 변수의 변화 알아보기 

for i in range(10):
    print("hello world", i)

```

```python
for imsi in range(0,10,2):
    print("hello world", imsi)
```

```python
for a in range(10,0):
    print("hello world", a)
```

```python
for i in range(5,12) :
    print("hello world", i)
```

```python
list(range(10,0))
```

```python
for a in range(10,0,-1):
    print("hello world", a)
```

```python
print(list(range(10)))
print(list(reversed(range(10))))
```

```python
for c in reversed(range(10)) : 
    print("hello world", c)
```

```python
# 시퀀스 객체로 반복하기

a = [10,20,30,40,50] # 리스트
for imsi in a :
    print(imsi)
```

```python
fruits = ("apple", "orange", "grape")

for fruit in fruits :
    print(fruit)
```

```python
for letter in "Python":
    print(letter)
```
