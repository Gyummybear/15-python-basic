```python
x = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40,
}
x.setdefault("e") # setdefault(변수) : 변수를 딕셔너라에 추가

print(x)
```

```python
# 딕셔너리에서 키의 값 수정하기 
x = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40,
}
```

```python
x.update(a = 90)

print(x)
```

```python
x.update(e=50)
print(x)
```

```python
x.update(a=900, f = 60) #  update로 여러개 키 값 변경 가능
print(x)
```

```python
y = {
    1 : "one",
    2 : "two"
}

y.update({1 : "ONE", 2 : "TWO"})  #  .update 딕셔너리 값 형태로도 작성가능
print(y)
```

```python
y.update(1: "하나")
print(y)
```

```python
y.update(1 ="하나")
print(y)
```

```python
y.update([[2,"TWO"], [4, "FOUR"]])
print(y)
```

```python
y.update(zip([1,2], ["one", "two"]))
print(y)
```

```python
# 딕셔너리에 할당과 복사 
x = {"a" : 0 , "b" : 0, "c" : 0, "d":0}

y = x
```

```python
x is y
print(id(x))
print(id(y))
```

```python
y["a"] = 99
print(x)
print(y)
```

```python
x = {"a" : 0 , "b" : 0, "c" : 0, "d":0}
y = x.copy()

x is y

y['a'] = 99

print(x)
print(y)
```

```python
x = {"a" : {"python" : "2.7"}, "b" : {"python" : "3.6"}}
y = x.copy()

y["a"]["python"] = "2.7.15"

print(x)
print(y)
```

```python
import copy

x = {"a" : {"python" : "2.7"}, "b" : {"python" : "3.6"}}
y = copy.deepcopy(x)

y["a"]["python"] = "2.7.15"

print(x)
print(y)
```
