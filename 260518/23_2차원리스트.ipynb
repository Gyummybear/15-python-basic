```python
# 반복문으로 2차원 리스트 요소를 모두 출력하기
# 2차원 리스트를 만들고 요소에 접근하기

a = [
    
        [10, 20], #3행 2열?
        [30, 40],
        [50, 60]
    
    ]

for x, y in a :
    print( x, y)
```

```python
a = [
    
        [10, 20], #3행 2열?
        [30, 40],
        [50, 60]
    
    ]

for i in a : 
    for j in i :
        print(j, end = " ")
    print()
```

```python
# for와 range , 사용하기

a = [
    
        [10, 20], #3행 2열?
        [30, 40],
        [50, 60]
    
    ]

for i in range(a): # a는 3
    
    for j in range(len(a[i])):
        print(a[i][j], end =  " ")
    print()
```

```python
# while 반복문을 한 번 사용하기

# for와 range , 사용하기

a = [
    
        [10, 20], #3행 2열? 
        [30, 40],
        [50, 60]
    
    ]

i = 0
while i < len(a):
    x , y = a[i]
    print(x,y)
    i += 1 
```

```python
a = [
    
        [10, 20], #3행 2열? 
        [30, 40],
        [50, 60]
    
    ]

i = 0
while i < len(a) : 

    j = 0
    while j < len(a[i]):
        print(a[i],[j], end = " ")
        j += 1
    print()
    i += 1
```

```python
#톱니형 리스트 만들기
a = [3,1,3,2,5]
b = []

for i in a : # a값 : 3,1,3,2,5
    line =  []
    for j in range(i): # i는 각각의 요소 (3,1,3,2,5)
         line.append(0) # line 0을 세 번 반복?
    b.append(line)
print(b)
```

```python
# 2차원 리스트의 할당과 복사 알아보기 
a = [
    [10, 20],
    [30, 40]
]

b = a

b[0][0] = 500 # 10값에 500을 집어넣음

print(a)
print(b)
```

```python
a = [
    [10, 20],
    [30, 40]
]

b = a.copy()
b[0][0] = 500
print(a)
print(b)
```

```python
x = [1,2,3,4]
y = x

y[0] = 10
print(x)
print(y)
```

```python
x = [1,2,3,4]
y = x.copy() # 여기서 [1,2,3,4] 값을 y으로 만듦

y[0] = 10  # y의 인덱스 값 0을 10으로 만듦
print(x)
print(y) # 그래서 변경 된 값이 y 값으로 나오는 것임
```
