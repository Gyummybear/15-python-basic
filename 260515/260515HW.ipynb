```python
a = int(input("첫 번째 값"))
b = int(input("두 번째 값"))

for i in range(a,b-1, -1) :
    if i % 3 == 0:
        print("Fizz")
    elif i % 7 == 0:
        print("Buzz")
    elif i % 21 == 0:
        print("FizzBuzz")
    else :
        print(i)
```

```python
s = int(input("역삼각형 높이를 입력하세요"))

for i in range(s) :
    for j in range(i):
        print("", end = "")

    for j in (range(2 * s - 1) - (2 * i)):
        print("*" , end = " ")
print()
```

```python
s = int(input("역삼각형 높이를 입력하세요: "))

for i in range(s):
    # 1. 공백 출력 (큰따옴표 안에 스페이스바 한 칸 필수)
    for j in range(i):
        print(" ", end="")

    # 2. 별 출력 (계산식을 range 괄호 안에서 한 번에 처리)
    for j in range((2 * s - 1) - (2 * i)):
        print("*", end="")
        
    # 3. 한 행이 끝날 때마다 줄바꿈 (첫 번째 for문 내부로 들여쓰기 일치)
    print()
```

```python
a = int(input(" 첫 번쨰 정수 "))
b = int(input(" 두 번째 정수 "))

result = []

for i in range(a,b+1):
    j = i ** 0.5
    k = round(j,1)

    result.append(k)
print(result)
```

```python
a = int(input(" 첫 번쨰 정수 "))
b = int(input(" 두 번째 정수 "))

result = [round(i**0.5 , 1) for i in range(a, b+1)]

print(result)
```
