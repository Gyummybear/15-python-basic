```python
# 리스트의 가장 작은 수, 가장 큰 수, 합계 구하기 
a = [38, 21, 53, 62, 19]

smallest = a[0]

for i in a :
    if i < smallest :
        smallest = i

print(smallest)
```

```python
a = [38, 21, 53, 62, 19]

largest = a[0]

for i in a :
    if i > largest:
        largest = i

print(largest)
```

```python
# 리스트 컴프리헨션
## 1. [식 for 변수 in 리스트] 
## 2. list(식 for 변수 in 리스트)

a = [i for i in range(10) ] 
print(a)

for i in range(10):
    print(i)


b = list(j for j in range(10))
print(b)
```

```python
# 리스트 표현에서 if 조건문 사용하기 

## 1. [식 for 변수 in 리스트 if 조건삭]
## 2. list(식 for 변수 in 리스트 if 조건식)

a = [i for i in range(10) if i % 2== 0 ]
print(a)
```

```python
b = [ a + 5 for a in range(10) if a % 2 == 1 ]

print(b)
```

```python
# for 반복문과 if 조건문을  여려 번 사용하기
# for 변수1 in 리스트1 if 조건식 1 
# for 변수 1 리스트2 if 조건식 2 

a = [
    i * j for j in range(2,10)
    for i in range(1,10)

]

print(a)
```
