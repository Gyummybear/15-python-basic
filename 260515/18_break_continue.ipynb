```python
# break로 반복문 끝내기

i = 0
while True :
    print(i)
    i += 1

    if i == 100:
        break
```

```python
# for에서 break로 반복문 끝내기

for i in range(10000):
    print(i)
    if i == 100 :
        break
```

```python
# continue 코드 실행 건너뛰기

for i in range (100):
    if i % 2 == 0:
        continue # 해당 되는 결과값 건너뛰고 다른 숫자 계산 (홀수만 출력)
    print(i)
```

```python
# whlie 반복문에서 continue로 코드 실행 건너뛰기

i = 0

while i < 100 :
    i += 1
    if i % 2 == 0 :
        continue
    print(i) #  if 안에 들여쓰기를 해야 if 구문을 실행한  i가 출력됨
```

```python
# 입력한 횟수 대로 반복하기

count = int(input("숫자"))

i = 0 
while True : 
    print(i)
    i += 1
    if count == i :
        break
```

```python
# 입력한 숫자까지 홀수 출력하기

count = int(input("숫자"))


for i  in range(count + 1) : # +1 한 이유는 입력을 안하면 숫자보다 하나 작게 나오기 때문에 마지막에 +1을 해야 입력한 숫자까지 출력됨
    if i % 2== 0:
        continue
    print(i)
```
