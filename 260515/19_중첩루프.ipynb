```python
# 계단식으로 별 출력하기

for i in range(5):
    for j in range(5): 
        if j <= i : # i 가 무조건 커야 실행이 되기 떄문에 , i = 1, 2, 3... 이 증가해야 j랑 최소 1씩 차이가 나야 *이 찍힘
            print("*", end = "")
    print()
```

```python
# 대각선으로 별 출력하기 
for i in range(5):
    for j in range(5):
        if j == i :
            print("*", end = "")
        else:
            print(" ", end= "")
    print()
```

```python
for i in range(5):
    for j in range(5):
        if i < j:
            print(' ' ,end = '')
        else :
            print('*', end = '')
    print()
```
