```python
# 오각형 만들기

import turtle as t

n = 6
t.shape("turtle")
for i in range(n):
    t.forward(100)
    t.right(360/n)
```

```python
# 다각형에 색칠하기
# 오각형 만들기

import turtle as t

n = 6
t.shape("turtle")
t.color("red")
t.begin_fill()
for i in range(5):
    t.forward(100)
    t.right(360/n)
t.end_fill()
```
